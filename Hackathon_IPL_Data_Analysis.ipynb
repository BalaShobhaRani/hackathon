# The column which has the most missing values according to the data is:


```python
missing_values_counts = df1.isnull().sum()

# the column with the maximum missing values
column_with_most_missing_values = missing_values_counts.idxmax()

print("Column with the most missing values:", column_with_most_missing_values)

# Display missing values counts for all columns
print("\nMissing values counts for all columns:")
print(missing_values_counts)
```

City which hosted most number of seasons according to the data is:


```python

city_season_counts = df1.groupby(['city', 'season']).size().reset_index(name='season_count')

# the city with the maximum count
max_season_city = city_season_counts.groupby('city')['season_count'].sum().idxmax()

print("City hosting the most number of seasons:", max_season_city)

# Display top 5 cities along with their counts
top_cities = city_season_counts.groupby('city')['season_count'].sum().nlargest(5).reset_index()
print("\nTop 5 cities and their season counts:")
print(top_cities)
```

What is the percentage of getting bat and field in feature 'toss_descision'?

```python
toss_decision_counts = df1['toss_decision'].value_counts()

# Calculate the percentage
percentage_bat = (toss_decision_counts['bat'] / toss_decision_counts.sum()) * 100
percentage_field = (toss_decision_counts['field'] / toss_decision_counts.sum()) * 100

print("Percentage of choosing to bat:", percentage_bat, "%")
print("Percentage of choosing to field:", percentage_field, "%")
```

  Which statement holds true

```python
field_wins = df1[df1['toss_decision'] == 'field']['winner'].count()
bat_wins = df1[df1['toss_decision'] == 'bat']['winner'].count()

print("Matches won by teams choosing to field:", field_wins)
print("Matches won by teams choosing to bat:", bat_wins)

if field_wins > bat_wins:
    print("Teams winning tosses and electing to field first have won more matches.")
elif field_wins < bat_wins:
    print("Teams winning tosses and electing to bat first have won more matches.")
else:
    print("Both A and B are equal.")
```

 To plot the most consistent batsmen among top 10 run getters, what plotting technique we use

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Plotting a box plot for the 'runs' column
sns.boxplot(x='batsman', y='batsman_runs', data=df2, order=df2['batsman'].value_counts().index[:10])

# Adding labels and title
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Box Plot of Runs for Top 10 Batsmen')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

In 2019 which batsman scored highest number of runs by hitting 6's and 4's?

```python

batsman_2019 = merged_df[merged_df['season'] == 2019]

# Calculate the total runs by hitting 6's and 4's for each batsman in 2019
batsman_runs_6_4 = batsman_2019.groupby('batsman')['batsman_runs'].sum()

# Identify the batsman with the highest runs by hitting 6's and 4's
top_batsman_6_4 = batsman_runs_6_4.idxmax()
runs_by_6_4 = batsman_runs_6_4.max()

print("In 2019, the batsman who scored the highest runs by hitting 6's and 4's is:", top_batsman_6_4)
print("Runs scored by hitting 6's and 4's:", runs_by_6_4)
```

```python
# Assuming df1 and df2 are your DataFrames
common_columns = ['match_id', 'batting_team', 'bowling_team']  # Specify actual common columns

# Merge based on common columns
joined_df = df1[common_columns+['win_by_wickets','season']].merge(df2[common_columns + ['batsman','batsman_runs','ball', 'bowler','non_striker','extra_runs']], on=common_columns, how='inner')
joined_df
```

 Most number of wickets taken by a bowler is

```python
top_bowlers = joined_df.groupby('bowler')['win_by_wickets'].sum().nlargest(5)

print("Top 5 bowlers based on total wickets:")
print(top_bowlers)
```

What is the strike rate of Kohli in 2016


```python
kohli_2016_stats = joined_df[(joined_df['batsman'] == 'V Kohli') & (joined_df['season'] == 2016)]

total_runs_kohli_2016 = kohli_2016_stats['batsman_runs'].sum()
total_balls_faced_kohli_2016 = kohli_2016_stats['ball'].count()

strike_rate_kohli_2016 = (total_runs_kohli_2016 / total_balls_faced_kohli_2016) * 100

print("Virat Kohli's strike rate in 2016:", strike_rate_kohli_2016)
```

which player has the highest hard-hitting ability

```python
hard_hitting_ability = df2.groupby('batsman').agg({'batsman_runs': 'sum', 'ball': 'count'})
hard_hitting_ability['strike_rate'] = (hard_hitting_ability['batsman_runs'] / hard_hitting_ability['ball']) * 100

players = ['AB de Villiers', 'CH Gayle', 'MS Dhoni', 'V Kohli']
highest_strike_rate_player = hard_hitting_ability.loc[players, 'strike_rate'].idxmax()

print("Player with the highest hard-hitting ability:", highest_strike_rate_player)
```
